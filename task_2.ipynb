```python
from bs4 import BeautifulSoup
import pandas as pd
import requests

headers = requests.utils.default_headers()
headers.update({ 'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/69.0'})
```

```python
from bs4 import BeautifulSoup
url = "https://tlgrm.ru/channels"

req = requests.get(url, headers)
soup = BeautifulSoup(req.content, 'html.parser')
classes = soup.find_all('a',attrs = {'class':'channel-category'})
```

```python
chennels_df = pd.DataFrame()

for class_a in classes:
    chennel_df = pd.DataFrame()
    class_name = class_a.text
    href_added = class_a.attrs['data-topic']
    class_url = url+'/'+href_added
    
    
    req = requests.get(class_url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    
    chennels = soup.find_all('div',attrs = {'class':'channel-card__info'})[1:]
    chennel_df['hrefs'] = [chennel.a.attrs['href'] for chennel in chennels]
    chennel_df['name'] = [chennel.a.text for chennel in chennels]
    chennel_df['class'] = href_added
    chennels_df = pd.concat([chennels_df,chennel_df], axis = 0)
```

```python
chennels_df.to_csv('data/chennels_df.csv', index = False)
```

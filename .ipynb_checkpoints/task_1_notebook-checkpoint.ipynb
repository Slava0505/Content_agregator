```python
import pandas as pd
def parse_n_messages(channel_name, n, client):
    channel = client.get_entity(channel_name)
    posts = client(GetHistoryRequest(
    peer=channel,
    limit=n,
    offset_date=None,
    offset_id=0,
    max_id=0,
    min_id=0,
    add_offset=0,
    hash=0))
    
    post_texts = [post.message for post in posts.messages]
    
    return pd.DataFrame(zip(post_texts,posts.messages), columns = ['text', 'message'])
```
